In 1971, a linguist named Morris Swadesh published a list of words most resistant to cross-cultural change. The list, now referred to as the Swadesh list (there are two primary versions, of 100 and 200 words), has been used to reconstruct genealogical relationships among the languages in a language family.

AMNH colleagues, Dr. Whiteley and Dr. Ward Wheeler (Curator, Invertebrate Zoology) have been studying historical relationships among languages, beginning with Uto-Aztecan languages of Middle and North America, using genetic methods developed in the field of evolutionary biology.
```python
#Clean data
for i in df_languages.columns:
  if "Unnamed" in i: #if the column name contains the substring "Unnamed"
    #I am going to remove `i` because it represents each column name in df.columns
    df_languages = df_languages.drop([i], axis=1) #axis = 1 tells the function that we are performing the .drop() method across a column instead of a row
```

```python
df_languages.head()
```

```python
possible_words = list(df_languages.columns)
possible_words.remove("Language")
print(possible_words)


possible_languages = df_languages["Language"].tolist()
print(possible_languages)
```

```python
#Translator function

def translate(dataframe, language, englishWord):
  #Find a specific value that matches a row and a column: .loc[row value, column value]
  translation = df_languages.loc[df_languages["Language"] == language, englishWord]
  translation = translation.to_string(index=False)
  # print(translation) #Displays the output
  return translation.strip()

```
